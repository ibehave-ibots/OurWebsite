```python
from jinja2 import Environment
```

```python
env = Environment()
```

```python
env.filters.keys()
```

```python
template = """
hello
hi
{{ name | upper | first}}
"""

print(env.from_string(template).render(name='Nick'))
```

```python
def small(name):
    return name.lower()
```

```python
env.filters['small'] = small
```

```python
template = """
hello
hi
{{ name | small}}
"""

print(env.from_string(template).render(name='NICK'))
```
